##File Format

`constellationship.fab`

Format: `AAA N 111111 222222 222222 333333 111111 444444`, where AAA is the three-letter constellation abbreviation, N is the number of line segments, 111111 222222 is a single line segment, and there are N pairs. 111111 is by the [Hipparcos Catalogue](https://en.wikipedia.org/wiki/Hipparcos#Published_catalogues) (HIP) designation for the star.

````python
def get_segments():
    segments = {}
    c_segments = '../data/ha_rey/ha_rey_english/constellationship.fab'
    with open(c_segments) as f:
        for line in f:
            c_data = line.split()
            if(len(c_data)<3):
                continue
            abbrev = c_data[0]
            c_data.remove(abbrev)
            num_segments = c_data[0]
            c_data.remove(num_segments)
            for i in range(len(c_data)):
                c_data[i] = int(c_data[i])
            segments[abbrev.lower()]=int(num_segments),c_data
    return segments

def get_constellations():
    constellations = {}
    c_names = '../data/ha_rey/ha_rey_english/constellation_names.eng.fab'
    segs = get_segments()
    with open(c_names) as f:
        for line in f:
            abbrev,name,temp=line.split('\t')
            abbrev = abbrev.lower()
            constellations[abbrev]=(name.strip('"'),segs[abbrev][0],segs[abbrev][1])
    return constellations

def get_stars():
    stars = {}
    s_names = '../data/ha_rey/ha_rey_english/star_names.fab'
    with open(s_names) as f:
        for line in f:
            hip,name=line.split('|')
            hip = int(hip.strip())
            stars[hip]=name.strip()
    return stars
````

````python
rey_asterisms = get_constellations()
print('Gemeni, labeled as {} by H. A. Rey has {} line segments.'.format(rey_asterisms['gem'][0],rey_asterisms['gem'][1]))
````

Only so many stars have [proper names](https://en.wikipedia.org/wiki/List_of_proper_names_of_stars), so it is to be expected that the `star_names.fab` list for the Rey asterisms doesn't have proper names for each star, and [PyEphem](https://github.com/brandon-rhodes/pyephem) only has 94 proper-named stars (filtered from [hip_main.dat.gz](ftp://adc.gsfc.nasa.gov/pub/adc/archives/catalogs/1/1239/hip_main.dat.gz)), so we'll want to gather the info from a catalog. In the [mini-hipparcos](https://github.com/digitalvapor/asterisms/blob/master/notebooks/mini-hipparcos.ipynb) notebook I found that there are 792 stars in the Rey asterisms. I chose to use the [hip2.dat](https://pystaratlas.googlecode.com/files/hip2.dat) catalog from [pystaratlas](https://code.google.com/p/pystaratlas/) and found out it was missing [HIP 55203](http://simbad.u-strasbg.fr/simbad/sim-id?Ident=HIP+55203) (Xi Ursae Majoris), so we'll manually add it. I also minified the hip2.dat into the [mini_rey_hip.dat](https://github.com/digitalvapor/asterisms/blob/master/data/mini_rey_hip.dat).

````python
# Usage:
# if(mini_hip(star)): print('Found star.')
# print(mini_hip(star,['RArad','DErad']))
# print(mini_hip(star,'all'))
def mini_hip(number,fields=None):
    field_names = ['HIP','Sn','So','Nc','RArad','DErad','Plx',
                   'pmRA','pmDE','e_RArad','e_DErad','e_Plx',
                   'e_pmRA','e_pmDE','Ntr','F2','F1','var','ic',
                   'Hpmag','e_Hpmag','sHp','VA','B_V','e_B_V',
                   'V_I','UW1','UW2','UW3','UW4','UW5','UW6',
                   'UW7','UW8','UW9','UW10','UW11','UW12','UW13',
                   'UW14','UW15']
    field_dict = {}
    i=0
    for field_name in field_names:
        field_dict[field_name]=i
        i += 1
    catalog = '../data/mini_rey_hip.dat'
    returned_values = []
    found = False
    with open(catalog) as f:
        for line in f:
            if(line.startswith('#')):
                continue
            star_data = line.split()
            # warn here or deal with if other than new line
            if(len(star_data)<3):
                continue
            if(int(star_data[0])==number):
                found = True
                if(fields=='all'):
                    return line
                elif(fields!=None and len(fields)>0):
                    for field in fields:
                        try:
                            returned_values.append(int(star_data[field_dict[field]]))
                        except:
                            returned_values.append(float(star_data[field_dict[field]]))
                    return returned_values
    return found

star_plot_dict = {}

for star in star_list:
    star = int(star)
    star_plot_dict[star] = mini_hip(star,['RArad','DErad'])
    try:
        temp_ra = star_plot_dict[star][0]
        temp_dec = star_plot_dict[star][1]
    except:
        temp_ra = None
        temp_dec = None
    try:
        print('HIP {} {:10} {:.3f} {:.3f}'.format(star,'('+s[star]+')',temp_ra,temp_dec))
    except:
        try:
            print('HIP {} {:10} {:.3f} {:5.3f}'.format(star,'',temp_ra,temp_dec))
        except:
            print('HIP {}'.format(star))
````

This example uses [shapely](https://github.com/Toblerity/Shapely). `pip install shapely`

````python
from matplotlib import pyplot
from shapely.geometry import LineString,Point

BLUE = '#6699cc'
GRAY = '#999999'

def plot_line(ax, ob):
    x, y = ob.xy
    ax.plot(x, y, color=BLUE, linewidth=3, solid_capstyle='round', zorder=1)

def plot_pt(ax, ob):
    x, y = ob.xy
    ax.plot(x, y, 'o',color=GRAY, zorder=1)
    
def display_constellation(name):
    tau = 2. * pi
    degree = tau / 360.
    hour = tau / 24.
    consts = get_constellations()
    star_segments = []
    #print(consts[name])
    fig = pyplot.figure(1,figsize(12,5),dpi=180)
    ax = fig.add_subplot(121)
    # plot segments
    for i in range(0,len(consts[name][2]),2):
        star1 = consts[name][2][i]
        star2 = consts[name][2][i+1]
        try:
            pt1 = tuple(mini_hip(star1,['RArad','DErad']))
            pt1 = (pt1[0]/hour,pt1[1]/degree)
            pt2 = tuple(mini_hip(star2,['RArad','DErad']))
            pt2 = (pt2[0]/hour,pt2[1]/degree)
            line = LineString([pt1,pt2])
            plot_line(ax,line)
        except:
            print('Segment ({},{}) needs fixin.'.format(star1,star2))
    stars = list(set(consts[name][2]))
    for i in range(len(stars)):
        try:
            pt = tuple(mini_hip(stars[i],['RArad','DErad']))
            pt = Point((pt[0]/hour,pt[1]/degree))
            plot_pt(ax,pt)
        except:
            print('Missing star {}.'.format(stars[i]))
    gca().invert_xaxis()
    gca().xaxis.grid(True)
    gca().yaxis.grid(True)
    ax.set_title(consts[name][0])
    #savefig('gemeni.png')

display_constellation('gem')
print('Here is Gemeni holding hands.')
````
