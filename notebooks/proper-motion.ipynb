If we are going to roll back time for constellations, we'll need to consider the proper motion of the stars.

Reference: http://www.unc.edu/~marzuola/Math547_S14/Projects/M_Jones.pdf

```python
import pylab
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def alphaUM(time):
    originalX = -61.3
    velocityX = -12.9
    originalY = 46.7
    velocityY = -20.5
    originalZ = 96.7
    velocityZ = -9.65
    KMtoLY = 1.05702341*10**-13
    newX = originalX + velocityX*KMtoLY*time
    newY = originalY + velocityY*KMtoLY*time
    newZ = originalZ + velocityZ*KMtoLY*time
    fig = pylab.figure()
    ax = Axes3D(fig)
    ax.plot([originalX], [originalY], [originalZ], 'd')
    ax.plot([newX], [newY], [newZ], 'd')
    pylab.axis('equal')
    #fig.savefig('alphaUM.png')
    return newX, newY, newZ
```

```python
alphaUM(100000)
```
